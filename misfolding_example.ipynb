```python
from PrototypeDB.orm import base
from PrototypeDB.orm.components import *
from PrototypeDB.orm.data import *
from IPython.display import HTML
import pandas as pd
import numpy as np

ome = base.Session()
```

```python
df = pd.DataFrame(index=[x.name for x in ome.query(ArrayExperiment).all()])
for gene_name in misfolding_genes:
    data = ome.query(GenomeData).join(Gene,ArrayExperiment).filter(Gene.name == gene_name).all()
    df[gene_name] = pd.DataFrame.from_dict({x.data_set.name:x.value for x in data}, orient='index')
HTML(df.dropna().sort(['yajL']).to_html())
```
